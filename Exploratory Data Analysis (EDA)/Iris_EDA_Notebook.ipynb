```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Iris.csv")
df.drop(columns=["Id"], inplace=True)
df.head()
```

## 1. Distribution of Numerical Features (Histograms)

```python

numerical_features = df.select_dtypes(include=["float64", "int64"]).columns

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)
    plt.hist(df[col], bins=20)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```


### Insights
- Petal features show clearer distributions
- Sepal features overlap more


## 2. Categorical Feature Analysis (Count Plot)

```python

sns.countplot(x="Species", data=df)
plt.title("Species Count")
plt.show()
```


### Insights
- Dataset is balanced
- No class imbalance issue


## 3. Outlier Detection (Box Plots)

```python

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```


### Insights
- Minor outliers in SepalWidth
- Petal features are stable


## 4. Correlation Heatmap

```python

plt.figure(figsize=(8, 6))
sns.heatmap(df[numerical_features].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```


### Insights
- Strong correlation between Petal Length and Petal Width
- Indicates multicollinearity
